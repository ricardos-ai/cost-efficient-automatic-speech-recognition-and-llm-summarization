## qwen2

Chunking text with overlapping to ensure that context between them is shared

```python
def chunk_text(text, max_chunk_size=3000, overlap=900):
    chunks = []
    words = text.split()
    start = 0

    while start < len(words):
        end = min(start + max_chunk_size, len(words))
        chunk = " ".join(words[start:end])
        chunks.append(chunk)
        start += max_chunk_size - overlap  # Move forward but keep overlap

    return chunks
```

```python
transcript_chunks = chunk_text(text)
transcript_chunks
```

```python
prompt_1 = """
You're an AI with medical knowledge, and you're asked to read a text from a video and answer the questions that are asked to you.
Act like your input is a medical case. Please respond in plain text without using Markdown formatting or special characters for headings, bullet points, code blocks, or links.

Examine the raw transcription of the medical operation video. While formulating your responses, never mention the transcript. Speak as if your information comes directly from the video recording of the operation. Ensure that all details remain true to the original transcription without altering or simplifying the language. Preserve all medical terminology as it appears in the transcript, ensuring fidelity to the original wording.

Key Instructions:
- Avoid repeating information that has already been provided in a previous chunk or step.
- Do not repeat headers like "Procedure Name & Purpose" unless that information is explicitly detailed in the chunk you are processing.
- If a section is missing information (e.g., no anesthesia used, no complications, etc.), leave it out of the response. Do not include default responses like "Not specified."
- The goal is to maintain clarity and avoid redundancy while summarizing the procedure.
- Extract and summarize only relevant information directly from the transcript. Exclude unrelated conversations.
- Ensure that the structure is logical, cohesive, and follows the progression of the procedure.

Structured Summary Format:

Procedure Name & Purpose:
[Only include if the specific procedure name and purpose are mentioned in the chunk]

Patient positioning:
[Include only if the patient's positioning is detailed]

Anesthesia used:
[Include only if mentioned]

Pre-procedure imaging/tests:
[Include only if mentioned]

Step-by-Step Summary of Operation:
[Only include steps mentioned in the chunk, avoid repeating from previous steps]

Tools & Techniques Used:
[Only include tools used in the current chunk, avoid repetition]

Medications administered:
[Only include if mentioned in this chunk]

Complications & Intraoperative Interventions:
[Only include if any complications or interventions are detailed]

Postoperative Actions & Final Status:
[Only include if mentioned in this chunk]
"""
```

```python
chunk_summaries = []

for i, chunk in enumerate(transcript_chunks):
    print(f"Processing chunk {i+1}/{len(transcript_chunks)}...")

    prompt_chunk = prompt_1 + chunk
    payload = {
        "model": "qwen2:7b",
        "prompt": prompt_chunk,
        "stream": False
    }

    response = requests.post(OLLAMA_ENDPOINT, json=payload)

    if response.status_code == 200:
        result = response.json()
        chunk_summary = result.get("response", "").strip()
        chunk_summaries.append(chunk_summary)
    else:
        print(f"Error processing chunk {i+1}: {response.status_code}, {response.text}")

print("Chunk Summaries:", chunk_summaries)
```

```python
final_prompt = """
You're an AI with medical knowledge, and you're asked to read a text from a video and answer the questions that are asked to you. Act like your input is a medical case.

Please respond only in plain text without using Markdown formatting or special characters for headings, bullet points, code blocks, or links. Examine the raw transcription of the medical operation video. While formulating your responses, never mention the transcript. Speak as if your information comes directly from the video recording of the operation.

Ensure that all details remain true to the original transcription without altering or simplifying the language. Preserve all medical terminology as it appears in the transcript, ensuring fidelity to the original wording. Do not make inferences or assumptions based on incomplete details. Ensure that the original language is fully preserved.

Now, consolidate the following summaries from different parts of the procedure while ensuring completeness and removing redundancy:

""" + "\n\n".join(chunk_summaries) + """

Instructions:
1. Consolidate all unique details from the summaries into a single final structured summary.
2. Remove redundancy: If a step is mentioned multiple times with different phrasing, merge them into one detailed mention.
3. Ensure completeness: Preserve every unique step, tool, technique, or complication while avoiding repetition.
4. Avoid paraphrasing unnecessarily: The original wording should be kept unless rewording is needed to remove redundancy.
5. Maintain logical flow: Organize the final summary naturally, ensuring a clear understanding of the procedure.

Please provide the final structured summary:
"""

payload = {
    "model": "qwen2:7b",
    "prompt": final_prompt,
    "stream": False
}

response = requests.post(OLLAMA_ENDPOINT, json=payload)

if response.status_code == 200:
    final_summary = response.json().get("response", "").strip()
    print("Final Structured Summary:\n")
    print(final_summary)
else:
    print("Error generating final summary:", response.status_code, response.text)
```

The structure of the text is good and the context is caught correctly, however, there are things that are said too vaguely and may be hallucinations. Therefore, a professional is needed - a human in the loop, who speaks the original tractsprict language in order to check 1) the translation of the text, 2) the summary.

And you can find difference between how many attention to details and mentioning them after two rounds of text. I think the reason is how the model uses context. If a large text is given, they pay more attention to certain parts than others or forget some parts of the text altogether. However, when we ask for summary of chunks, the context is preserved more, since the "concentration" of each original text in the chunks in Sammari is greater, and then one pass completes the text of a much smaller volume, so the level of detail is better than when sending the full text at once.
